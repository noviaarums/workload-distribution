```python
# NUMBER OF COMMITS THROUGH OUT THE PROJECT HISTORY

import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'scrapy_commits.csv'
data = pd.read_csv(file_path)

# Calculate the number of unique committers
num_committers = data['Committer'].nunique()
print(f"Total number of all committers: {num_committers}")

# Calculate the total number of commits (count the number of rows)
num_commits = len(data)
print(f"Total number of all commits: {num_commits}")

# Group by committer then count the occurences
all_committer = data.groupby(['Committer']).size()

# Plot the distribution of all committer
all_committer.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Committer')
plt.ylabel('Number of Commits')
plt.title('Distribution of Committers and Their Commits')
# plt.legend(title='Number of Commits')
plt.xticks(rotation=45)

# Show the Plot the distribution of all reviewer and state
plt.tight_layout()
plt.show()

```

```python
# TOP 20 COMMITTERS WITH HIGHEST COMMITS COUNT THROUGH OUT THE PROJECT HISTORY
# AND ITS PERCENTAGE OF ALL TOTAL NUMBER COMMITS

import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'scrapy_commits.csv'
data = pd.read_csv(file_path)

# Calculate total commits per committer
total_commits = data.groupby('Committer').size()

# Get the top 20 commiters
top_20_committers = total_commits.nlargest(20)

# Calculate the percentage of commits by each top 20 committer
top_20_percentage = (top_20_committers / total_commits.sum() ) * 100


# Combine into a DataFrame
top20committers_df = pd.DataFrame ({'Number of Commits': top_20_committers}).sort_values(by='Number of Commits', ascending=False)
top20percentage_df = pd.DataFrame ({'Number of Commits Percentage': top_20_percentage}).sort_values(by='Number of Commits Percentage', ascending=False)

# Create subplots
fig, axes = plt.subplots(2,1, figsize=(12, 6))

# Plot 1: Total Reviews
top20committers_df.plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Top 20 Committers by Number of Commits')
axes[0].set_xlabel('Committer')
axes[0].set_ylabel('Number of Commits')
axes[0].tick_params(axis='x', rotation=45)

# Annotate each bar
for i, value in enumerate(top20committers_df['Number of Commits']):
    axes[0].text(
        i,  # x-coordinate
        value + 0.5,  # y-coordinate (slightly above the bar)
        f'{value}',  # formatted value
        ha='center',  # horizontal alignment
        va='bottom',  # vertical alignment
        fontsize=8  # font size
    )

# Plot 2: Percentage of Total Reviews
top20percentage_df.plot(kind='bar', ax=axes[1], color='lightcoral')
axes[1].set_title('Top 20 Committers by Percentage of Total Commits')
axes[1].set_xlabel('Committer')
axes[1].set_ylabel('Percentage of Total Commits')
axes[1].tick_params(axis='x', rotation=45)

# Annotate each bar with the percentage value
for i, value in enumerate(top20percentage_df['Number of Commits Percentage']):
    axes[1].text(
        i,  # x-coordinate
        (value + 1),  # y-coordinate (slightly above the bar)
        f'{value:.2f}%',  # formatted value
        ha='center',  # horizontal alignment
        va='bottom',  # vertical alignment
        fontsize=8  # font size
    )

# Show the plots
plt.tight_layout()
plt.show()

```

```python
# NUMBER OF COMMITS IN THE LAST YEAR (2024)
# ITS PERCENTAGE OF ALL COMMITS HAPPENED IN THE LAST YEAR

import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'scrapy_commits.csv'
data = pd.read_csv(file_path)

# Extract the year from the `Date of Committed` column
data['year'] = data['Date of Committed'].str[:4]  # Get the first 4 characters as the year

# Filter data for the year 2024
data_2024 = data[data['year'] == '2024']

# Group by committer and count the occurrences for 2024
commits_2024 = data_2024.groupby(['Committer']).size()

# Calculate the percentage of commits in 2024
commitsPer_2024 = (commits_2024 / commits_2024.sum() ) * 100

# Make into a DataFrame
commits_2024_df = pd.DataFrame ({'Number of Commits': commits_2024}).sort_values(by='Number of Commits', ascending=False)
commitsPer_2024_df = pd.DataFrame ({'Number of Commits Percentage': commitsPer_2024}).sort_values(by='Number of Commits Percentage', ascending=False)

# Print the number of unique committers
num_committers_2024 = len(commits_2024)
print(f"Number of unique committers in 2024: {num_committers_2024}")

# Print the total number of commits in 2024
total_commits_2024 = commits_2024.sum()
print(f"Total number of commits in 2024: {total_commits_2024}")

# Get the name of the committer with the highest number of commits
top_committer = commits_2024.idxmax()
top_commits = commits_2024.max()
print(f"The committer with the highest number of commits in 2024 is: {top_committer} with {top_commits} commits.")


# Create subplots
fig, axes = plt.subplots(2,1, figsize=(12, 6))

# Plot 1: Total Commits
commits_2024_df.plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Committers by Number of Commits in 2024')
axes[0].set_xlabel('Committer')
axes[0].set_ylabel('Number of Commits')
axes[0].tick_params(axis='x', rotation=45, labelsize=6)

# Annotate each bar
for i, value in enumerate(commits_2024_df['Number of Commits']):
    axes[0].text(
        i,  # x-coordinate
        value + 0.5,  # y-coordinate (slightly above the bar)
        f'{value}',  # formatted value
        ha='center',  # horizontal alignment
        va='bottom',  # vertical alignment
        fontsize=6  # font size
    )

# Plot 2: Percentage of Total Committers
commitsPer_2024_df.plot(kind='bar', ax=axes[1], color='lightcoral')
axes[1].set_title('Committers by Percentage of Total Commits in 2024')
axes[1].set_xlabel('Committer')
axes[1].set_ylabel('Percentage of Total Commiters')
axes[1].tick_params(axis='x', rotation=45, labelsize=6)

# Annotate each bar with the percentage value
for i, value in enumerate(commitsPer_2024_df['Number of Commits Percentage']):
    axes[1].text(
        i,  # x-coordinate
        (value + 1),  # y-coordinate (slightly above the bar)
        f'{value:.2f}%',  # formatted value
        ha='center',  # horizontal alignment
        va='bottom',  # vertical alignment
        fontsize=6,  # font size
        rotation=75
    )

# Show the plots
plt.tight_layout()
plt.show()
```
