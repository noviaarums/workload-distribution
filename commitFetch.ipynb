```python
from github import Github

# github Token
g = Github("Replace this with your own github token")
```

```python
import pandas as pd

count_repo = 1
count_repo_stop = 1

#10 projects
#1. scrapy/scrapy
#2. fastapi/fastapi
#3. psf/requests --- keras-team/keras xxx
#4. pallets/flask
#5. celery/celery
#6. python-pillow/Pillow
#7. python-poetry/poetry
#8. psf/black
#9. pydantic/pydantic
#10. encode/django-rest-framework
projects = ["psf/requests"]

for repo_name in projects:
  repo = g.get_repo(repo_name)
  print(f"Analyzing repository: {repo.full_name}")

  #Fetch pull requests
  pulls = repo.get_pulls(state='closed', sort='created', direction='desc')

  c_list = []

  for pull in pulls:
    commits = pull.get_commits() #Get commits on each pull request
    for commit in commits:
      try:
        c_list.append ({
          'ID Commit' : commit.sha,
          'ID Author' : commit.author.id,
          'Author' : commit.author.login,
          'Date of Authored' : commit.commit.author.date,
          'ID Committer' : commit.committer.id,
          'Committer' : commit.committer.login,
          'Date of Committed' : commit.commit.committer.date
        })
        print(f"ID Commit : {commit.sha}, ID Author : {commit.author.id}, Author : {commit.author.login}, Date of Authored : {commit.commit.author.date}, ID Committer : {commit.committer.id}, Committer : {commit.committer.login}, Date of Committed : {commit.commit.committer.date}")
      except:
        print("error")

  df = pd.DataFrame(c_list)
  if not df.empty:
    df.to_csv(f'{repo.name}_commits.csv', index=False)

  if count_repo == count_repo_stop:
    break
  count_repo += 1
```
