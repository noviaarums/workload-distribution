```python
from github import Github

# github Token
g = Github("Replace this with your own github token")
```

```python
import pandas as pd

count_repo = 1
count_repo_stop = 2

#10 projects
#1. scrapy/scrapy
#2. fastapi/fastapi
#3. psf/requests
#4. pallets/flask
#5. celery/celery
#6. python-pillow/Pillow
#7. python-poetry/poetry
#8. psf/black
#9. pydantic/pydantic
#10. encode/django-rest-framework
projects = ['pydantic/pydantic','psf/requests']

for repo_name in projects:
  repo = g.get_repo(repo_name)
  print(f"Analyzing repository: {repo.full_name}")
  df = pd.DataFrame(columns=['Reviewer', 'State', 'Submitted At'])

  #Fetch pull requests
  pulls = repo.get_pulls(state='closed', sort='created', direction='desc')

  a_list = []

  for pull in pulls:
    reviews = pull.get_reviews() #Get reviews on each pull request
    for review in reviews:
      try:
        a_list.append ({
          'ID Review' : review.id,
          'Submitted At': review.submitted_at,
          'ID Reviewer': review.user.id,
          'Reviewer': review.user.login,
          'State': review.state,
          'ID Commit': review.commit_id
        })
        print(f"ID Review : {review.id}, Submitted At: {review.submitted_at}, ID Reviewer: {review.user.id}, Reviewer: {review.user.login}, State: {review.state}, ID Commit: {review.commit_id}")
      except:
        print("error")

  df = pd.DataFrame(a_list)
  if not df.empty:
    df.to_csv(f'{repo.name}_reviews.csv', index=False)

  if count_repo == count_repo_stop:
    break
  count_repo += 1
```
